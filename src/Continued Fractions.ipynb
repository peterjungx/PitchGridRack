```python

def i2c(i):
    x = i%11
    y = i//11
    return (y + x%6 + x//6, 2*y + x//6)

exquis_coords=[i2c(i) for i in range(61)]

def c2i(c):
    u,v = c
    return u+v*6-2*(v//2)-v%2

[c2i(c)-i for i,c in enumerate(exquis_coords)]
```

```python
def gcd(a, b):
    while (b != 0):
        t = b
        b = a % b
        a = t
    return a

gcd(2,2)
```

```python
from fractions import Fraction

def continued_fraction(x, n=10):
    """Generate the continued fraction representation of x as a list."""
    cf = []
    while True:
        a = int(x)
        cf.append(a)
        if x == a:
            break
        x = 1 / (x - a)
        if x.is_integer() or len(cf) == n:
            break
    return cf

def convergents(cf):
    """Generate convergents from a continued fraction list."""
    p,q,p_next, q_next = 0,1, 1,0
    result = []
    for a in cf:
        p, q, p_next, q_next = p_next, q_next, a * p_next + p, a * q_next + q
        result.append(Fraction(p_next, q_next))
        
    return result

def closest_rational(x, n):
    """Find the closest rational approximation to x with deviation at most 1/n."""
    cf = continued_fraction(x)
    convs = convergents(cf)
    for frac in convs:
        if abs(x - float(frac)) <= 1/n:
            return frac
    # If no fraction is found within the tolerance, return the last fraction
    return convs[-1] if convs else None

# Example usage:
x = 1.414223532  # Approximation of sqrt(2)
n = 24  # Tolerance

approx = closest_rational(x, n)
print(f"Original number: {x}")
print(f"Approximation: {approx}")
print(f"Difference: {abs(x - float(approx))}")
print(f"Max allowed difference: {1/n}")
```

```python
import numpy as np
for x in np.linspace(0,12,1000):
    print(2**(x/12), closest_rational(2**(x/12), 12))

set([closest_rational(2**(x/12), 12) for x in np.linspace(0,12,10000)])
```
